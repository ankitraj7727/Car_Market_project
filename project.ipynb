```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('car_data.csv') # Replace with your exact file name

# Look at the first 5 rows
df.head()
```

```python
# Check for missing values
print("Missing values per column:\n", df.isnull().sum())

# Drop rows with missing values (or fill them using NumPy)
df = df.dropna()

# Check the data types
print("\nData Types:\n", df.dtypes)
```

```python
# What are the top 5 most expensive cars in this dataset?
top_expensive = df.nlargest(5, 'selling_price') # Change column name based on your dataset
print("Top 5 Most Expensive Cars:\n", top_expensive[['name', 'selling_price']])

# Average price of cars by fuel type (Petrol vs Diesel vs Electric)
avg_price_fuel = df.groupby('fuel')['selling_price'].mean().round(2)
print("\nAverage Price by Fuel Type:\n", avg_price_fuel)
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['selling_price'], bins=30, kde=True, color='purple')
plt.title('Distribution of Car Selling Prices')
plt.xlabel('Price')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
# Using Seaborn to see if higher max_power means higher price
sns.scatterplot(data=df, x='km_driven', y='selling_price', hue='transmission', palette='deep')
plt.title('Kilometers Driven vs Selling Price')
plt.xlabel('Kilometers Driven')
plt.ylabel('Selling Price')
plt.show()
```
